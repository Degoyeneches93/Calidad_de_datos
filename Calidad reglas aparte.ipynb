```python
# REGLA DE COMPLETITUD - PY_COMP_1
def verificar_completitud(df_catalogo, columna):
    #Verifica la completitud de una columna en un dataframe.
    nulos = df_catalogo[columna].isnull().sum()
    vacios = (df_catalogo[columna] == '').sum()
    total_filas = len(df_catalogo)
    valores_validos = total_filas - (nulos + vacios)
    porcentaje_calidad = f"{(valores_validos / total_filas) * 100:.2f}%" if total_filas > 0 else "0.00%"

    return valores_validos, total_filas, porcentaje_calidad

# Almacenar resultados
resultados = []
columnas_con_errores = []

# Iterar sobre las configuraciones en df_REGLAS_FACT
for _, row in df_REGLAS_FACT.iterrows():
    columna = row['ATRIBUTO']
    codigo_regla = row['COD_REGLA']

    # Evaluar solo las reglas de completitud
    if codigo_regla == 'PY_COMP_1':
        valores_validos, total_filas, porcentaje_calidad = verificar_completitud(df_catalogo, columna)

        # Almacenar resultado en una lista
        resultados.append({
            'Atributo': columna,
            'Total Filas Evaluadas': total_filas,
            'Valores Válidos': valores_validos,
            'Porcentaje Calidad': porcentaje_calidad
        })

        # Verificar si la calidad no es 100%
        if porcentaje_calidad != "100.00%":
            columnas_con_errores.append(columna)

# Crear un DataFrame con los resultados
df_resultados = pd.DataFrame(resultados)

# Crear un DataFrame con los rechazos
df_errores = pd.DataFrame(columns=['Código del Canal', 'Canal', 'Atributo', 'Dato'])
for columna in columnas_con_errores:
    df_error_col = df_catalogo[df_catalogo[columna].isnull() | (df_catalogo[columna] == '')][['Código del Canal', 'Canal']].copy()
    df_error_col['Dato'] = df_catalogo[columna]
    df_error_col['Atributo'] = columna
    df_errores = pd.concat([df_errores, df_error_col], ignore_index=True)


# Exporta el dataframe de resultados a un archivo CSV en una ruta específica
#df_resultados.to_csv(r'D:\Usuarios\1015437134\OneDrive - Banco Popular\Escritorio\EG\Calidad\Catálogo de Canales\resultados_calidad.csv', index=False, encoding='utf-8')

# Exportar el dataframe de rechazos a un archivo CSV
#df_errores.to_csv(r'D:\Usuarios\1015437134\OneDrive - Banco Popular\Escritorio\EG\Calidad\Catálogo de Canales\errores_calidad.csv', index=False, encoding='utf-8')
# Mostrar resultados
#print(df_resultados)
print(df_errores)
#print(resultados)
```

```python
#REGLA DE PRECISIÓN PY_PREC_1
# Definir las relaciones entre columnas del catálogo y las columnas del dataframe consolidado
relaciones = {
    'Código Familia del Canal': 'Código Familia del Canal',  
    'Familia del Canal': 'Familia del Canal',
    'Código Clase del Canal': 'Código Clase del Canal',
    'Clase del Canal': 'Clase del Canal',
    'Código Subclase del Canal': 'Código Subclase del Canal',
    'Subclase del Canal': 'Subclase del Canal',
    'Código del Canal': 'Código del Canal',
    'Canal': 'Canal'
}

# Función para validar múltiples columnas contra un dataframe consolidado
def verificar_precision_consolidada(df_catalogo, df_listaref_JQ_FCSC, relaciones):
    resultados_generales = []
    df_rechazos_consolidado = pd.DataFrame()


    for col_df1, col_df2 in relaciones.items():
        # Verificar coincidencias entre las columnas
        df_merge = df_catalogo.merge(df_listaref_JQ_FCSC[[col_df2]], left_on=col_df1, right_on=col_df2, how='left', indicator=True)
        total_filas = len(df_catalogo)
        df_rechazos = df_merge[df_merge['_merge'] != 'both'].copy()
        valores_validos = total_filas - len(df_rechazos)
        porcentaje_calidad = f"{(valores_validos / total_filas) * 100:.2f}%" if total_filas > 0 else "0.00%"

        # Agregar resultados generales
        resultados_generales.append({
            'Atributo': col_df1,
            'Total Filas Evaluadas': total_filas,
            'Rechazos': len(df_rechazos),
            'Porcentaje Calidad': porcentaje_calidad
        })
        
        if len(df_rechazos) > 0:
            # Generar dataframe de rechazos para este atributo
            df_rechazos_resultado = df_rechazos[[col_df1, 'Código del Canal', 'Canal']].copy()
            df_rechazos_resultado['Atributo'] = col_df1
            df_rechazos_resultado.rename(columns={col_df1: 'Dato'}, inplace=True)

            # Reordenar las columnas
            df_rechazos_resultado = df_rechazos_resultado[['Código del Canal', 'Canal', 'Atributo', 'Dato']]

            # Consolidar rechazos
            df_rechazos_consolidado = pd.concat([df_rechazos_consolidado, df_rechazos_resultado], ignore_index=True)


    return resultados_generales, df_rechazos_consolidado, 


# Ejecutar la función para validar
resultados_generales, df_rechazos = verificar_precision_consolidada(df_catalogo, df_listaref_JQ_FCSC, relaciones)

# Crear el dataframe con el indicador de calidad general
df_resultado_general = pd.DataFrame(resultados_generales)

# Exportar resultados generales
#df_resultado_general.to_csv(r'D:\EG\Calidad\Catálogo de Canales\resultados_calidad_validez_multiple.csv', index=False, encoding='utf-8-sig')

# Exportar rechazos si existen
#if not df_rechazos.empty:
 #   df_rechazos.to_csv(r'D:\EG\Calidad\Catálogo de Canales\rechazos_validez_multiple.csv', index=False, encoding='utf-8-sig')
#else:
 #   print("No se generó el archivo de rechazos porque no hubo datos rechazados.")
#print(df_resultado_general)
print(df_rechazos)

```

```python
# REGLA DE VALIDEZ - LONGITUD VARIOS ATRIBUTOS PY_VALZ_1
# Función para validar la longitud de una columna específica
def validar_longitud(df_catalogo, columna, longitud_deseada):
    # Identificar rechazos basados en la longitud
    rechazos = df_catalogo[df_catalogo[columna].str.len() != longitud_deseada] 
    # Identificar registros rechazados por nulos o vacíos
    rechazos_nulos_vacios = df_catalogo[
        (df_catalogo[columna].isnull()) | (df_catalogo[columna] == '')
    ]
    # Concatenar ambos tipos de rechazos
    rechazos = pd.concat([rechazos, rechazos_nulos_vacios]).drop_duplicates()
    # Calcular métricas de calidad
    total_filas = len(df_catalogo)
    total_rechazos = len(rechazos)
    valores_validos = total_filas - total_rechazos
    porcentaje_calidad = f"{((total_filas - total_rechazos) / total_filas) * 100:.2f}%" if total_filas > 0 else "0.00%"
   
    return {
        'atributo': columna,
        'total_filas': total_filas,
        'rechazos': total_rechazos,
        'porcentaje_calidad': porcentaje_calidad,
        'valores validos': valores_validos
    }, rechazos

# Definir las columnas a verificar y las longitudes deseadas
longitudes_columnas = {
    'Código Familia del Canal': 2,
    'Código Clase del Canal': 4,
    'Código Subclase del Canal': 9,
    'Código del Canal': 11
}

# Almacenar los resultados en una lista
resultados = []
detalles_rechazos = pd.DataFrame()

for columna, longitud_deseada in longitudes_columnas.items():
    # Ejecutar la función de validación de longitud para cada columna con su longitud respectiva
    resultado, rechazos = validar_longitud(df_catalogo, columna, longitud_deseada)
    resultados.append(resultado)
    
    # Si hay rechazos, añadir información relevante
    if not rechazos.empty:
        rechazos['Atributo'] = columna
        rechazos = rechazos[['Atributo', columna, 'Código del Canal', 'Canal']]
        rechazos.rename(columns={columna: 'Dato'}, inplace=True)

        # Reiniciar los índices de rechazos y resetear también columnas duplicadas si existen
        rechazos = rechazos.loc[:, ~rechazos.columns.duplicated()].copy()
        rechazos.reset_index(drop=True, inplace=True)
        
        # Concatenar los rechazos sin problemas de índice
        detalles_rechazos = pd.concat([detalles_rechazos, rechazos], ignore_index=True)

# Convertir los resultados en un DataFrame para el archivo de calidad general
df_resultados = pd.DataFrame(resultados)

# Guardar el archivo de resultados generales
#ruta_guardado_resultados = r'D:\Usuarios\1015437134\OneDrive - Banco Popular\Escritorio\EG\Calidad\Catálogo de Canales\resultados_calidad_longitud.csv'
#df_resultados.to_csv(ruta_guardado_resultados, index=False)

# Guardar el archivo con los rechazos
#ruta_guardado_rechazos = r'D:\Usuarios\1015437134\OneDrive - Banco Popular\Escritorio\EG\Calidad\Catálogo de Canales\rechazos_longitud.csv' 
#detalles_rechazos.to_csv(ruta_guardado_rechazos, index=False)

#print(f"El archivo de resultados generales se guardó en: {ruta_guardado_resultados}")
#print(f"El archivo de rechazos se guardó en: {ruta_guardado_rechazos}")
print(detalles_rechazos)
```

```python
# REGLA DE VALIDEZ -  TIPO DE DATO PY_VALZ_2
# Función para validar el tipo de contenido de una columna específica
def validar_contenido(df_catalogo, columna, tipo_contenido):
    if tipo_contenido == 'numerico':
        es_valido = df_catalogo[columna].apply(lambda x: str(x).isdigit() if pd.notnull(x) else False)
    elif tipo_contenido == 'alfanumerico':
        es_valido = df_catalogo[columna].apply(
            lambda x: all(c.isalnum() or c.isspace() or c in "*.,;:-/()" for c in str(x)) if pd.notnull(x) else False)
    elif tipo_contenido == 'texto':
        es_valido = df_catalogo[columna].apply(
        lambda x: all(c.isalpha() or c.isspace() or c in "*.,;:-/()" for c in str(x)) if pd.notnull(x) else False)
    else:
        raise ValueError(f"Tipo de contenido {tipo_contenido} no soportado.")

    # Identificar registros rechazados
    rechazos = df_catalogo[~es_valido]

    # Calcular métricas de calidad
    total_filas = len(df_catalogo)
    total_rechazos = len(rechazos)
    valores_validos = total_filas - total_rechazos
    porcentaje_calidad = f"{(valores_validos / total_filas) * 100:.2f}%" if total_filas > 0 else "0.00%"

    return {
        'atributo': columna,
        'total_filas': total_filas,
        'rechazos': total_rechazos,
        'porcentaje_calidad': porcentaje_calidad,
        'valores_validos': valores_validos
    }, rechazos

# Configuración de columnas a verificar y tipos de contenido
contenido_columnas = {
    'Código Familia del Canal': 'numerico',
    'Familia del Canal': 'texto',
    'Código Clase del Canal': 'numerico',
    'Clase del Canal': 'texto',
    'Código Subclase del Canal': 'numerico',
    'Subclase del Canal': 'texto',
    'Código del Canal': 'numerico',
    'Canal': 'texto',
    'Transacciones del canal' : 'alfanumerico',
    'Factor de autenticación del canal' : 'alfanumerico',
    'Estado del canal' : 'texto',
    'Tipo de atención del canal' : 'texto',
    'ID Superfinanciera': 'numerico',
    'Grupo del canal': 'texto'
}

# Almacenar los resultados en una lista y los rechazos en un DataFrame
resultados = []
detalles_rechazos = pd.DataFrame()

for columna, tipo_contenido in contenido_columnas.items():
    # Ejecutar la función de validación para cada columna con su tipo de contenido
    resultado, rechazos = validar_contenido(df_catalogo, columna, tipo_contenido)
    resultados.append(resultado)

    # Si hay rechazos, añadir información relevante
    if not rechazos.empty:
        rechazos['Atributo'] = columna
        rechazos = rechazos[['Atributo', columna, 'Código del Canal', 'Canal']]
        rechazos.rename(columns={columna: 'Dato'}, inplace=True)

        # Eliminar columnas duplicadas
        rechazos = rechazos.loc[:, ~rechazos.columns.duplicated()].copy()

        # Reiniciar los índices de rechazos y asegurarse de que sean únicos
        rechazos = rechazos.reset_index(drop=True)

        # Concatenar los rechazos sin problemas de índice
        detalles_rechazos = pd.concat([detalles_rechazos, rechazos], ignore_index=True)

# Ordenar detalles_rechazos según el orden especificado
detalles_rechazos = detalles_rechazos[['Código del Canal', 'Canal', 'Atributo', 'Dato']]

# Convertir los resultados en un DataFrame para el archivo de calidad general
df_resultados = pd.DataFrame(resultados)


# Guardar los resultados y rechazos en archivos específicos
#ruta_resultados = r'./resultados_calidad_contenido.csv'
#ruta_rechazos = r'./rechazos_contenido.csv'

#df_resultados.to_csv(ruta_resultados, index=False)
#detalles_rechazos.to_csv(ruta_rechazos, index=False)

#print(f"El archivo de resultados generales se guardó en: {ruta_resultados}")
#print(f"El archivo de rechazos se guardó en: {ruta_rechazos}")
print(detalles_rechazos)
```

```python
#CAMPO NUM_DIM_FECHA_SK - df_resultado_final
# Convertir la columna 'DTM_DIM_TIE_DIA_NK' a formato datetime si no lo está ya
df_DIM_TIEMPO['DTM_DIM_TIE_DIA_NK'] = pd.to_datetime(df_DIM_TIEMPO['DTM_DIM_TIE_DIA_NK'], errors='coerce')
# Definir la fecha en que se corre la calidad
fecha_requerida = datetime.now().replace(hour=0, minute=0, second=0, microsecond=0)
# Buscar el valor de NUM_DIM_TIEMPO_SK para la fecha específica
num_dim_tiempo_sk = df_DIM_TIEMPO.loc[df_DIM_TIEMPO['DTM_DIM_TIE_DIA_NK'] == fecha_requerida, 'NUM_DIM_TIEMPO_SK'].values
# Verificar si se encontró el valor y mostrar el resultado
if len(num_dim_tiempo_sk) > 0:
    num_dim_tiempo_sk = num_dim_tiempo_sk[0]
    #print(f"El valor de NUM_DIM_TIEMPO_SK para la fecha {fecha_requerida.strftime('%d-%m-%Y')} es: {num_dim_tiempo_sk}")
else:
    raise ValueError("Fecha no encontrada en df_DIM_TIEMPO.")

#EDITABLE ¡¡
#CAMPO NUM_DIM_DOMINIO_SK
# Definir el nombre del dominio que deseas buscar
nombre_dominio = "Catálogo de Canales"  # Cambiar el nombre del dominio sobre el que se realiza la calidad
# Buscar el valor de NUM_DIM_DOMINIO_SK para el nombre del dominio específico
num_dim_dominio_sk = df_DIM_DOMINIO.loc[df_DIM_DOMINIO['STR_NOMBRE_DOMINIO'] == nombre_dominio, 'NUM_DIM_DOMINIO_SK'].values
# Verificar si se encontró el valor y mostrar el resultado
if len(num_dim_dominio_sk) > 0:
    num_dim_dominio_sk = num_dim_dominio_sk[0]
    #print(f"El valor de NUM_DIM_DOMINIO_SK para el dominio '{nombre_dominio}' es: {num_dim_dominio_sk}")
else:
    raise ValueError("Dominio no encontrado en df_DIM_DOMINIO.")

# Definir las funciones de validación

def verificar_completitud(df_catalogo, columna):
    total_filas = len(df_catalogo)
    valores_validos = df_catalogo[columna].notnull().sum()
    return valores_validos, total_filas

def verificar_precision_consolidada(df_catalogo, df_listaref, relaciones, columna):
    # Ajustar para validar solo la columna especificada
    if columna in relaciones:
        valores_validos = df_catalogo[columna].isin(df_listaref[relaciones[columna]]).sum()
    else:
        raise ValueError(f"La columna {columna} no está definida en las relaciones.")
    total_filas = len(df_catalogo)
    return valores_validos, total_filas

def validar_longitud(df_catalogo, columna, longitud_deseada):
    total_filas = len(df_catalogo)
    valores_validos = df_catalogo[columna].apply(lambda x: len(str(x)) == longitud_deseada if pd.notnull(x) else False).sum()
    return valores_validos, total_filas

def validar_contenido(df_catalogo, columna, tipo_contenido):
    if tipo_contenido == 'numerico':
        es_valido = df_catalogo[columna].apply(lambda x: str(x).isdigit() if pd.notnull(x) else False)
    elif tipo_contenido == 'alfanumerico':
        es_valido = df_catalogo[columna].apply(
            lambda x: all(c.isalnum() or c.isspace() or c in "*.,;:-/()" for c in str(x)) if pd.notnull(x) else False)
    elif tipo_contenido == 'texto':
        es_valido = df_catalogo[columna].apply(
            lambda x: all(c.isalpha() or c.isspace() or c in "*.,;:-/()" for c in str(x)) if pd.notnull(x) else False)
    else:
        raise ValueError(f"Tipo de contenido {tipo_contenido} no soportado.")

    # Identificar registros rechazados
    rechazos = df_catalogo[~es_valido]
    total_filas = len(df_catalogo)
    total_rechazos = len(rechazos)
    valores_validos = total_filas - total_rechazos
    return valores_validos, total_filas
     
# Definir el DataFrame final
df_resultado_final = pd.DataFrame()

# Iterar sobre las configuraciones en df_REGLAS_FACT
for _, row in df_REGLAS_FACT.iterrows():
    # Extraer valores de configuración
    columna = row['ATRIBUTO']
    codigo_dato_mgi = row['CÓDIGO_DATO_MGI']
    codigo_regla = row['COD_REGLA']

    # Determinar valores válidos y filas totales en función del tipo de regla
    if codigo_regla == 'PY_COMP_1':
        valores_validos, total_filas = verificar_completitud(df_catalogo, columna)
    elif codigo_regla == 'PY_PREC_1':
        valores_validos, total_filas = verificar_precision_consolidada(df_catalogo, df_listaref_JQ_FCSC, relaciones, columna)
    elif codigo_regla == 'PY_VALZ_1':
        longitud_deseada = longitudes_columnas.get(columna, None)  # Obtener longitud deseada si está definida
        if longitud_deseada:
            valores_validos, total_filas = validar_longitud(df_catalogo, columna, longitud_deseada)
        else:
            raise ValueError(f"No se encuentra longitud deseada para la columna {columna}.")
    elif codigo_regla == 'PY_VALZ_2':  # Nueva regla para tipo de contenido
    # Determinar tipo de contenido según la lógica ya definida
        tipo_contenido = 'numerico' if 'ID' in columna else 'texto'
        if 'Canal' in columna:
            tipo_contenido = 'alfanumerico'  # Ejemplo lógico para determinar el tipo de contenido
        valores_validos, total_filas = validar_contenido(df_catalogo, columna, tipo_contenido)
    else:
        raise NotImplementedError(f"La regla {codigo_regla} no está implementada.")

    # Depuración detallada
    #print(f"DEBUG: Procesando columna={columna}, regla={codigo_regla}")
    #print(f"Valores válidos: {valores_validos}, Total filas: {total_filas}")
    #print(f"Datos ejemplo de la columna:")
    #print(df_catalogo[columna].head())

    # Validar que valores_validos y total_filas sean consistentes
    if valores_validos > total_filas:
        print(f"ERROR DETECTADO: columna={columna}, valores_validos={valores_validos}, total_filas={total_filas}")
        raise ValueError(f"Los valores válidos superan el total de filas para la columna {columna}.")
    
    # Obtener valores de las tablas DIM
    num_dim_dato_sk = df_DIM_DATO.loc[df_DIM_DATO['SRT_CODIGO_DATO_MGI'] == codigo_dato_mgi, 'NUM_DIM_DATO_SK'].values[0]
    num_dim_regla_sk = row['NUM_DIM_REGLA_SK']
    num_codigo_dimension = df_REGLAS.loc[df_REGLAS['COD_REGLA'] == codigo_regla, 'NUM_DIMENSION'].values[0]
    nombre_dimension = df_REGLAS.loc[df_REGLAS['COD_REGLA'] == codigo_regla, 'NOM_DIMENSION'].values[0]
    niveles_por_atributo = df_REGLAS_FACT.groupby('ATRIBUTO')['NUM_NIVEL'].max().reset_index()
    dimensiones_por_atributo = df_REGLAS_FACT.groupby('ATRIBUTO')['NUM_DIMENSION'].nunique().reset_index()

    # Crear DataFrame temporal para esta configuración
    df_temp = pd.DataFrame([{
        'NUM_DIM_FECHA_SK': num_dim_tiempo_sk,
        'NUM_DIM_DOMINIO_SK': num_dim_dominio_sk,
        'NUM_DIM_DATO_SK': num_dim_dato_sk,
        'NUM_DIM_REGLA_SK': num_dim_regla_sk,
        'STR_COD_CLASIFICACION': 0,
        'STR_DESC_CLASIFICACION': "Clasificacion General",
        'NUM_CODIGO_DIMENSION': num_codigo_dimension,
        'STR_NOMBRE_DIMENSION': nombre_dimension,
        'NUM_CODIGO_FUENTE': 46,
        'STR_NOMBRE_FUENTE': "ARCHIVO USUARIO",
        'NUM_TOTAL_REGIS_OK_REGLA': valores_validos,
        'NUM_TOTAL_REGIS_EVALUADOS': total_filas,
        'DTM_FCHA_CREACION': fecha_requerida
    }])
    
    # Aplicar cálculos adicionales como NUM_POR_CALIDAD_X_DIMEN, NUM_PESO, etc.
    df_temp['NUM_POR_CALIDAD_X_DIMEN'] = (df_temp['NUM_TOTAL_REGIS_OK_REGLA'] / df_temp['NUM_TOTAL_REGIS_EVALUADOS']).fillna(0)


# CALCULO DEL CAMPO NUM_POR_CALIDAD_X_DATO
    # Paso 1: Realizar un merge para asociar NUM_DIM_DATO_SK de df_REGLAS_FACT con NUM_POR_CALIDAD_X_DIMEN de df_temp
    df_merged = df_REGLAS_FACT.merge(
        df_temp[['NUM_DIM_DATO_SK', 'NUM_POR_CALIDAD_X_DIMEN']],
        on='NUM_DIM_DATO_SK',
        how='inner'
    )

    # Paso 2: Sumar los valores de NUM_POR_CALIDAD_X_DIMEN agrupados por NUM_DIM_DATO_SK
    suma_por_calidad = (
        df_merged.groupby('NUM_DIM_DATO_SK', as_index=False)['NUM_POR_CALIDAD_X_DIMEN']
        .sum()
        .rename(columns={'NUM_POR_CALIDAD_X_DIMEN': 'suma_calidad'})
    )

    # Depuración: Verificar los resultados
    print("Suma de NUM_POR_CALIDAD_X_DIMEN agrupada por NUM_DIM_DATO_SK:")
    print(suma_por_calidad)



    # Paso 2: Contar las reglas únicas (NUM_DIM_REGLA_SK) aplicadas a cada NUM_DIM_DATO_SK
    conteo_reglas = (
        df_REGLAS_FACT.groupby('NUM_DIM_DATO_SK')['NUM_DIM_REGLA_SK']
        .nunique()
        .reset_index(name='conteo_reglas')
    )

    # Paso 3: Combinar los resultados de suma y conteo
    calidad_por_dato = suma_por_calidad.merge(conteo_reglas, on='NUM_DIM_DATO_SK')

    # Paso 4: Calcular NUM_POR_CALIDAD_X_DATO dividiendo suma_calidad entre conteo_reglas
    calidad_por_dato['NUM_POR_CALIDAD_X_DATO'] = calidad_por_dato['suma_calidad'] / calidad_por_dato['conteo_reglas']

    # Paso 5: Mapear el resultado al DataFrame principal df_temp
    df_temp = df_temp.merge(
        calidad_por_dato[['NUM_DIM_DATO_SK', 'NUM_POR_CALIDAD_X_DATO']],
        on='NUM_DIM_DATO_SK',
        how='left'
    )

    # Depuración: Verificar los resultados
    print("Resultados del cálculo de NUM_POR_CALIDAD_X_DATO (agrupados):")
    print(calidad_por_dato)
    print("df_temp actualizado con NUM_POR_CALIDAD_X_DATO:")
    print(df_temp[['NUM_DIM_DATO_SK', 'NUM_POR_CALIDAD_X_DIMEN', 'NUM_POR_CALIDAD_X_DATO']])




    max_nivel_por_dato = niveles_por_atributo.loc[niveles_por_atributo['ATRIBUTO'] == columna, 'NUM_NIVEL'].values[0]
    num_dimensiones_diferentes = dimensiones_por_atributo.loc[dimensiones_por_atributo['ATRIBUTO'] == columna, 'NUM_DIMENSION'].values[0]
    df_temp['NUM_PESO'] = max_nivel_por_dato * num_dimensiones_diferentes

#NOTA NUM_PESO_REGIS_EVALUADOS / NUM_PESO_REGIS_OK_REGLA: Si hay más de una regla evaluada de una sola dimensión, el valor se divide 
# entre la cantidad de reglas de dicha dimensión. Por ejemplo si el atributo tiene 4 reglas de validez, el resultado de 
# NUM_PESO_REGIS_EVALUADOS debe dividirse por 4 en esos registros
    # Paso previo: Calcular NUM_PESO_REGIS_EVALUADOS y NUM_PESO_REGIS_OK_REGLA
    df_temp['NUM_PESO_REGIS_EVALUADOS'] = df_temp['NUM_TOTAL_REGIS_EVALUADOS'] * max_nivel_por_dato
    df_temp['NUM_PESO_REGIS_OK_REGLA'] = df_temp['NUM_TOTAL_REGIS_OK_REGLA'] * max_nivel_por_dato
    # Paso 1: Identificar las combinaciones de NUM_DIM_DATO_SK y STR_NOMBRE_DIMENSION con reglas repetidas
    conteo_dimensiones = (
        df_REGLAS_FACT.groupby(['NUM_DIM_DATO_SK', 'STR_NOMBRE_DIMENSION'])
        .size()
        .reset_index(name='conteo'))
    # Depuración: Verificar el conteo de dimensiones
    #print("Conteo de dimensiones por NUM_DIM_DATO_SK y STR_NOMBRE_DIMENSION:")
    #print(conteo_dimensiones)
    # Paso 2: Fusionar el conteo con df_temp
    df_temp = df_temp.merge(
        conteo_dimensiones,
        on=['NUM_DIM_DATO_SK', 'STR_NOMBRE_DIMENSION'],
        how='left')
    # Paso 3: Ajustar NUM_PESO_REGIS_EVALUADOS y NUM_PESO_REGIS_OK_REGLA para combinaciones con conteo > 1
    df_temp.loc[df_temp['conteo'] > 1, 'NUM_PESO_REGIS_EVALUADOS'] /= df_temp.loc[df_temp['conteo'] > 1, 'conteo']
    df_temp.loc[df_temp['conteo'] > 1, 'NUM_PESO_REGIS_OK_REGLA'] /= df_temp.loc[df_temp['conteo'] > 1, 'conteo']
    # Paso 4: Eliminar la columna 'conteo' para limpiar el DataFrame final
    df_temp.drop(columns=['conteo'], inplace=True)

    # Depuración: Verificar los valores ajustados
    #print("Valores ajustados de NUM_PESO_REGIS_EVALUADOS y NUM_PESO_REGIS_OK_REGLA para reglas repetidas:")
    #print(df_temp[['NUM_DIM_DATO_SK', 'STR_NOMBRE_DIMENSION', 'NUM_PESO_REGIS_EVALUADOS', 'NUM_PESO_REGIS_OK_REGLA']])





    df_temp['Reglas'] = df_temp['NUM_DIM_DOMINIO_SK'].astype(str) + \
                       df_temp['NUM_DIM_DATO_SK'].astype(str) + \
                       df_temp['NUM_DIM_REGLA_SK'].astype(str)

    # Combinar el resultado con el DataFrame final
    df_resultado_final = pd.concat([df_resultado_final, df_temp], ignore_index=True)

# Especifica la ruta donde deseas guardar el archivo
#ruta_exportacion = r"C:\Degoyeneches\Estudio\Python\Calidad de datos\FACT_CATALOGO_CANALES.xlsx"

# Exporta el DataFrame
#df_resultado_final.to_excel(ruta_exportacion, index=False, engine='openpyxl')

#print(f"Archivo exportado exitosamente en: {ruta_exportacion}")

# Mostrar el DataFrame consolidado con los resultados para todos los atributos

print(suma_por_calidad)
```
